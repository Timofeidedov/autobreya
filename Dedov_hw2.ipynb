```python
import pymorphy2
```

```python
morph = pymorphy2.MorphAnalyzer(lang='ru')
```

```python
f = open('corptext.txt','r')
```

```python
text=''
```

```python
for smth in f:
  text=text+smth+' '
```

```python
import string
```

```python
text_w_p = [i.strip(string.punctuation) for i in text.split()]
```

```python
text_w_p[0]='Кейсу'
```

```python
pymorphypars=[]
```

```python
for smth in text_w_p:
  pymorphypars.append(morph.parse(smth)[0].tag.POS)
```

```python
pymorphypars=list(filter((None).__ne__, pymorphypars))
```

```python
len(pymorphypars)
```

```python
pymorphypars_fixed=[]
```

```python
for smth in pymorphypars:
  if smth=='ADJF' or smth=='ADJS' or smth=='COMP':
    pymorphypars_fixed.append('ADJ')
  elif smth=='ADVB' or smth=='PRED':
    pymorphypars_fixed.append('ADV')
  elif smth=='NUMR':
    pymorphypars_fixed.append('NUM')
  elif smth=='NPRO':
    pymorphypars_fixed.append('PRON')
  elif smth=='NUMR':
    pymorphypars_fixed.append('NUM')
  elif smth=='PRCL':
    pymorphypars_fixed.append('PART')
  elif smth=='VERB' or smth=='INFN':
    pymorphypars_fixed.append('V')
  elif smth=='PRTF' or smth=='PRTS':
    pymorphypars_fixed.append('PCP')
  else:
    pymorphypars_fixed.append(smth)
```

```python
corpus["pymorphy"] = pymorphypars_fixed
```

```python
from deeppavlov import build_model, configs
```

```python
model = build_model(configs.morpho_tagger.UD2_0.morpho_ru_syntagrus_pymorphy, download=True)
```

```python
deeppars=[]
```

```python
for smth in text_w_p:
  print(smth)
  if smth!='':
    for parse in model([smth]):
      print(parse)
      deeppars.append(parse.split()[2])
```

```python
deeppars
```

```python
len(deeppars)
```

```python
deeppars_fixed=[]
```

```python
for smth in deeppars:
  if smth=='DET':
    deeppars_fixed.append('PRON')
  elif smth=='VERB' or smth=='AUX':
    deeppars_fixed.append('V')
  elif smth=='CCONJ' or smth=='SCONJ':
    deeppars_fixed.append('CONJ')
  elif smth == 'ADP':
    deeppars_fixed.append('PREP')
  else:
    deeppars_fixed.append(smth)
```

```python
corpus["deeppavlov"] = deeppars_fixed
```
