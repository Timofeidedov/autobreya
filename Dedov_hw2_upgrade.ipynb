```python
def chunker(model,text):
  chunkerlist=[]
  ngrams=[]
  for parse in model([text]):
    chunkerlist.append(parse)
  i=0
  while i<len(chunkerlist):
    if 'не' in chunkerlist[i] and 'ADJ' in chunkerlist[i+1]:
      ngrams.append(chunkerlist[i].split()[1]+' '+chunkerlist[i+1].split()[1])
      i+=2
    elif ("сомнительного" in chunkerlist[i] or "сомнительный" in chunkerlist[i]) and ('NOUN' in chunkerlist[i+1]):
      ngrams.append(chunkerlist[i].split()[1]+' '+chunkerlist[i+1].split()[1])
      i+=2
    elif 'ADP' in chunkerlist[i] and 'NOUN' in chunkerlist[i+1]:
      ngrams.append(chunkerlist[i].split()[1]+' '+chunkerlist[i+1].split()[1])
      i+=2
    else:
      ngrams.append(chunkerlist[i].split()[1])
      i+=1
  return(ngrams)
```
